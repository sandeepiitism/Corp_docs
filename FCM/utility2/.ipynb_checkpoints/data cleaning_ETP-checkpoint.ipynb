```python
import pandas as pd
import numpy as np

def clean_data(sheet_name):
    data = pd.read_excel("fcm_data_2.xlsx", skiprows=2, sheet_name=sheet_name, usecols="A:K")
    data = data[["Date", "Tank No.", "pH", "Cond ", "BOD", "COD ", "TOC", "Turb ", 
                    "Hardness"]]
    data['Date'].fillna(method='ffill', inplace=True)
    data['Tank No.'] = data['Tank No.'].replace(r'^After\s*$', np.nan, regex=True)
    data['Tank No.'].fillna(method='ffill', inplace=True)
    data = data.dropna(subset=['Hardness'])
    data['Tank No.'] = data['Tank No.'].str.replace(r'\s*\(Before\)', '', regex=True)
    data['Tank No.'] = data['Tank No.'].str.replace(' \( Before \)', '')
    return data

data_sheet_0 = clean_data(0)
data_sheet_1 = clean_data(1)
data_sheet_2 = clean_data(2)
data_sheet_3 = clean_data(3)
data_sheet_4 = clean_data(4)
data_sheet_5 = clean_data(5)
data_sheet_6 = clean_data(6)
data_sheet_7 = clean_data(7)
data_sheet_8 = clean_data(8)
data_sheet_9 = clean_data(9)
data_sheet_10 = clean_data(10)
```

```python
combined_data = pd.concat([data_sheet_0, data_sheet_1, data_sheet_2, data_sheet_3, data_sheet_4,
                          data_sheet_5, data_sheet_6, data_sheet_7, data_sheet_8, data_sheet_9, data_sheet_10])

# Save the combined data to a CSV file
combined_data.to_csv('cleaned_fcm.csv', index=False)
```

```python
import pandas as pd
import numpy as np

def clean_data(sheet_name):
    data = pd.read_excel("fcm_data_2.xlsx", skiprows=2, sheet_name=sheet_name, usecols="A:K")
    data = data[["Date", "Tank No.", "pH", "Cond ", "BOD", "COD ", "TOC", "Turb ", 
                    "Hardness"]]
    data['Date'].fillna(method='ffill', inplace=True)
    data['Tank No.'] = data['Tank No.'].replace(r'^After\s*$', np.nan, regex=True)
    data['Tank No.'].fillna(method='ffill', inplace=True)
    data = data.dropna(subset=['Hardness'])
    data['Tank No.'] = data['Tank No.'].str.replace(r'\s*\(Before\)', '', regex=True)
    data['Tank No.'] = data['Tank No.'].str.replace(' \( Before \)', '')
    # Convert 'Hardness' column to string type
    data['Hardness'] = data['Hardness'].astype(str)
    # Remove '+' sign from 'Hardness' values
    data['Hardness'] = data['Hardness'].str.replace('+', '')
    return data


data_sheet_0 = clean_data(0)
data_sheet_1 = clean_data(1)
data_sheet_2 = clean_data(2)
data_sheet_3 = clean_data(3)
data_sheet_4 = clean_data(4)
data_sheet_5 = clean_data(5)
data_sheet_6 = clean_data(6)
data_sheet_7 = clean_data(7)
data_sheet_8 = clean_data(8)
data_sheet_9 = clean_data(9)
data_sheet_10 = clean_data(10)

combined_data = pd.concat([data_sheet_0, data_sheet_1, data_sheet_2, data_sheet_3, data_sheet_4,
                          data_sheet_5, data_sheet_6, data_sheet_7, data_sheet_8, data_sheet_9, data_sheet_10])

# Save the combined data to a CSV file
combined_data.to_csv('cleaned_fcm_1.csv', index=False)
```
