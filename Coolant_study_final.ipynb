```python
import pandas as pd
import numpy as np
import pyodbc
import warnings

warnings.filterwarnings("ignore")
```

```python
import pyodbc
cnxn = pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                        "Server=na0vm00024.apac.bosch.com;"
                        "Database=DB_MFC2DB_SQL;"
                        "Trusted_Connection=yes;")
```

```python
sql = """SELECT [Coolant_Set_pressure] ,[CoolantPr], [PresentTool], [ToolBroken], [ProgRun], [TimeStamp] FROM [ChironDZ9_1] WHERE Timestamp BETWEEN '2023-09-20T00:00:47' AND '2024-02-21T09:45:58'
  ORDER BY 'TimeStamp'""" 
df = pd.read_sql_query(sql, cnxn, index_col="TimeStamp", parse_dates=True)
df.head()
```

```python
df.tail()
```

```python
df_copy = df[df["ToolBroken"] & df["ProgRun"]]
df_copy
```

## When Tool Broken

### Signatures of Tool 23 Coolant Pressure before break

```python
### Tool_23 ---> 26th Sept, 2023

import plotly.graph_objects as go

trace1 = go.Scatter(x=tool23_on_26sep.index, y=tool23_on_26sep['Coolant_Set_pressure'], mode='lines', name='Coolant_Set_pressure', line=dict(color='blue'))

trace2 = go.Scatter(x=tool23_on_26sep.index, y=tool23_on_26sep['CoolantPr'], mode='lines', name='CoolantPr', line=dict(color='red'))

annotations = []
for timestamp, coolant_pressure, present_tool in zip(tool23_on_26sep.index, tool23_on_26sep['CoolantPr'], tool23_on_26sep['PresentTool']):
    annotations.append(dict(x=timestamp, y=coolant_pressure, text=f'Tool {present_tool}', showarrow=False, xanchor='center', yanchor='bottom'))


layout = go.Layout(
    title='Coolant Set Pressure vs Coolant Pressure',
    xaxis=dict(title='TimeStamp'),
    yaxis=dict(title='Value'),
    annotations=annotations
)


fig = go.Figure(data=[trace1, trace2], layout=layout)


fig.show()
```
